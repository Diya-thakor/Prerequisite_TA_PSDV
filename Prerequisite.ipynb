## 1. List Operations and Basic Statistics:
- Create a list of the first 20 natural numbers.
- Compute the sum of all even numbers in the list.
- Compute the product of all odd numbers in the list.
- Calculate the mean, median, and standard deviation of the list.

```python
import statistics
# l1 = [];
# for i in range(1,21):
#   l1.append(i)

# One line approach:
l1 = list(range(1,21))

print("List of 1st 20 natural number: ",l1)

even_sum = 0
for ele in l1:
  if ele%2==0:
    even_sum += ele

# even_sum = sum(num for num in l1 if num % 2 == 0)
print("Sum of all even numbers: " , even_sum)

product = 1;
for ele in l1:
  if ele%2:
    product *= ele

# product = reduce(lambda x, y: x * y, (num for num in l1 if num % 2 != 0))

print("Product of all odd numbers: " , product)

mean = sum(l1)/len(l1)
print("Mean: " , mean)

sorted_l1 = sorted(l1)
n = len(sorted_l1)

if(n%2==0):
  median = (sorted_l1[n//2] + sorted_l1[n//2-1])/2
else:
  median = sorted_l1[n//2]

print("Median: " , median)

mean_diff_squares = [(x-mean) ** 2 for x in l1]
variance = sum(mean_diff_squares) / (len(l1)-1)
stdev = variance ** 0.5

print("Standard Deviation: " , stdev)

print("Using statistics module: ")
print("Mean: " , statistics.mean(l1))
print("Median: " , statistics.median(l1))
print("Standard Deviation: " , statistics.stdev(l1))
```

##3. Fibonacci Sequence:
- Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
- Plot the Fibonacci sequence using a line graph.

```python
import matplotlib.pyplot as plt

def fibonacci(n):
  if n==0:
    return []
  if n==1:
    return [0]
  if n==2:
    return [0,1]

  fib_list = [0,1]
  for i in range(2,n):
    fib_list.append(fib_list[i-1]+fib_list[i-2])

  return fib_list

n = 10
fib_list = fibonacci(n)
print(fib_list)

def plot_fibonacci(n):
    fib = fibonacci(n)
    plt.figure(figsize=(8, 5))
    plt.plot(range(n), fib, marker='o', linestyle='-', color='b')
    plt.title(f'Fibonacci Sequence (First {n} Numbers)', fontsize=14)
    plt.xlabel('Index (n)', fontsize=12)
    plt.ylabel('Fibonacci Number', fontsize=12)
    plt.grid(True)
    plt.show()
plot_fibonacci(n)
```

## 9. Merging Dictionaries:
- Write Python code to merge two dictionaries. If a key is present in both, sum their values.

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

merged_dict = {}

for key, value in dict1.items():
  if key in dict2:
    merged_dict[key] = value + dict2[key]
  else:
    merged_dict[key] = value

for key, value in dict2.items():
  if key not in merged_dict:
    merged_dict[key] = value

print(merged_dict)
```

## 10. Character Frequency:
- Write a function that accepts a string and:
- Converts all characters to lowercase.
- Counts the frequency of each character.
- Returns a dictionary with characters as keys and their counts as values.
- Visualize the character frequencies using a bar chart.

```python
import matplotlib.pyplot as plt

def get_character_frequency(string) :
    string = string.lower()
    char_freq = {}
    for char in string:
      if char in char_freq:
        char_freq[char] += 1
      else:
        char_freq[char] = 1
    return char_freq

string = input("Enter string: ")
char_freq = get_character_frequency(string)

print(char_freq)

plt.figure(figsize=(8, 5))
plt.bar(char_freq.keys(), char_freq.values(), color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Character')
plt.ylabel('Frequency')
plt.title('Character Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 11. Remove Duplicates:
- Write a program to remove duplicate elements from a list while maintaining the order of elements.
- Calculate the mean and median of the unique elements.

```python
import statistics

def remove_duplicates(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

input_list = [10, 20, 10, 30, 20, 40, 50, 30]
print("Original List:", input_list)

unique_elements = remove_duplicates(input_list)
print("Unique Elements:", unique_elements)

mean = statistics.mean(unique_elements)
median = statistics.median(unique_elements)

print("Mean of Unique Elements:", mean)
print("Median of Unique Elements:", median)
```

## 12. String Operations:
- Given a list of strings, write a Python function to:
 - Find the longest string.
 - Count how many strings start with a vowel.
 - Visualize the distribution of string lengths using a histogram.

```python
import matplotlib.pyplot as plt

def string_operation(string_list):
  longest_string = max(string_list,key=len)
  vowels = {'a','e','i','o','u'}
  vowel_count = sum(1 for string in string_list if string[0].lower() in vowels)

  string_lengths = [len(string) for string in string_list]

  return longest_string,vowel_count,string_lengths

# string_list=["hello","how","are","you","nice","worked","wrong","right","question"]
string_list = ["apple", "banana", "school", "umbrella", "cat", "elephant", "notebook", "owl", "dog"]
longest_string,vowel_count,string_lengths = string_operation(string_list)

print("String list: ",string_list)
print("Longest string of list : ", longest_string)
print("Strings start with a vowel: ",vowel_count)

plt.figure(figsize=(8,5))
plt.hist(string_lengths,bins=range(min(string_lengths),max(string_lengths)+2),color='skyblue',edgecolor='black',alpha=0.7)
plt.xlabel('String Length')
plt.ylabel('Frequency')
plt.title('Distribution of String Lengths')
plt.plot()
```

## 15. Anagram Finder:
- Write a Python function to find all anagrams of a word in a given list.
- Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
- Output: ["enlist", "inlets"]

```python
def get_anagrams(string, string_list):
  anagrams = []
  for word in string_list:
    if sorted(string) == sorted(word):
      anagrams.append(word)
  return anagrams

word = "listen"
string_list = ["enlist", "google", "inlets", "banana"]

anagrams = get_anagrams(word,string_list)
print("Anagrams of word 'listen' are ", anagrams)
```

## 21. Basic Statistics:
- Write Python code to calculate:
 - Mean, median, mode, and standard deviation of a list of numbers.
 - Visualize the data using a box plot.

```python
import matplotlib.pyplot as plt
import random

numbers = [random.randint(1,50) for _ in range(10)]
print("Numbers: ",numbers)

mean = sum(numbers)/len(numbers)
print("Mean: ",mean)

sorted_numbers = sorted(numbers)
if len(sorted_numbers)%2 == 0:
  median = (sorted_numbers[len(sorted_numbers)//2-1]+sorted_numbers[len(sorted_numbers)//2])/2
else:
  median = sorted_numbers[len(sorted_numbers)//2]

print("Median: ",median)

frequency = {}
for num in numbers:
  if num in frequency:
    frequency[num] += 1
  else:
    frequency[num] = 1

mode = max(frequency,key=frequency.get)
print("Mode: ",mode)

squared_diffs = [(x - mean) ** 2 for x in numbers]
variance = sum(squared_diffs) / len(numbers)
std_dev = variance ** 0.5
print("Standard Deviation: ",std_dev)

plt.figure(figsize=(8,5))
plt.boxplot(numbers)
plt.ylabel('Values')
plt.title('Box Plot of Numbers');
```

## 22. Normalization:
- Write a Python function to normalize a list of numbers to a range of 0 to 1.
- Visualize the normalized data using a line plot.

```python
import matplotlib.pyplot as plt
import random

def normalize(numbers):
  min_num = min(numbers)
  max_num = max(numbers)
  normalized_numbers = [(x - min_num) / (max_num - min_num) for x in numbers]
  return normalized_numbers

numbers = [random.randint(1,50) for _ in range(10)]
print("Numbers: ",numbers)

normalized_numbers = normalize(numbers)
print("Normalized numbers: ",normalized_numbers)

plt.figure(figsize=(8,5))
plt.plot(normalized_numbers,marker='o',linestyle='-',color='skyblue')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Normalized Data')
plt.grid(True)
```
